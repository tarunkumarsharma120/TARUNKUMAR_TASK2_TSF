# <font face='MV Boli'><font color=green><b> Importing the library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
%matplotlib inline
```

# <font face='MV Boli'><font color=green><b> Load the iris dataset

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head() 
```

# <font face='MV Boli'><font color=green><b> Finding the optimum number of clusters for k-means classification

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

# <font face='MV Boli'><font color=green><b> Plotting the results onto a line graph

```python
plt.grid
fig=plt.figure(facecolor='lightblue',figsize=(12,7),dpi=70,edgecolor='blue')
ax=fig.add_subplot(111)
ax.grid()
ax.plot(range(1, 11), wcss,c='orange',marker='*',mec='blue',ms=15)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

# <font face='MV Boli'><font color=green><b> Applying kmeans to the dataset / Creating the kmeans classifier

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

# <font face='MV Boli'><font color=green><b> Visualising the clusters & Plotting the centroids of the clusters

```python
fig=plt.figure(facecolor='pink',figsize=(12,7),dpi=70,edgecolor='blue')
ax=fig.add_subplot(111)
ax.grid()
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],s = 50, c = 'red', label = 'Iris-setosa',marker='*',)
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],s = 50, c = 'blue', label = 'Iris-versicolour',marker='o')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 50, c = 'green', label = 'Iris-virginica',marker='^')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'orange', label = 'Centroids')
plt.legend()
```
